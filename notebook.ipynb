# Movimiento Browniano

```python
# Importar librer√≠as
import random
import matplotlib.pyplot as pyplot
```

```python
# Declarar array
brownianMotion = []
```

```python
# Declarar variables
steps = 200
dist = 10
pos = 0
```

```python
# Simular
for n in range(steps):
    rand = random.randint(0,1)
    if rand == 0:
        pos = pos+dist
    else:
        pos = pos-dist
    brownianMotion.append(pos)

# Estilizar
pyplot.title("Movimiento Browniano")
pyplot.xlabel("Pasos")
pyplot.ylabel("Distancia")

# Graficar
pyplot.plot(brownianMotion)

pyplot.show()
```
